```python
def common_elements(list1,list2):
    res = []
    for num in list1:
        if num in list2 and num not in res:
            res.append(num)
    return res
l1 = [1,2,3,4]
l2 = [3,4,5,6]
print(common_elements(l1,l2))
```

```python
def words_count(sentence):
    sentence = sentence.lower()
    words = sentence.split()
    freq = {}
    for word in words:
        if word in words and word not in freq:
            freq[word] = 1
        else:
            freq[word] += 1
    return freq
sen = "The cat and The hat"
print(words_count(sen))
    
```
